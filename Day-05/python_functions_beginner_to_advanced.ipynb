# Python Functions

## What is a Function?

A function is a block of code that performs a specific task.

Instead of writing the same code again and again, we write it once inside a function and reuse it whenever needed.

## Why do we use Functions?

We use functions to:

- Avoid repeating the same code
- Make code clean and organized
- Reuse code multiple times
- Break a big problem into small parts
- Make programs easier to understand

## Simple Definition

A function means:

**Write once, use many times.**


# 1. Function Without Parameters

This is the simplest type of function.

It does not take any input.

It only runs the code written inside it.


````python
def login(username, password):
    if username == "user" and password == "123":
        return "Login Successful"
    else:
        return "Login Failed"
````

````python
def login(password):
    if password == "1234":
        print("Access Granted")
    else:
        print("Access Denied")
````

````python
def greet():
    print("Welcome to Python class")

# greet()
````

# Explanation

```python
def greet():
```

- `def` is used to create a function
- `greet` is the function name
- `()` means this function has no input
- `:` means the function block starts
- Indentation is required inside the function

```python
greet()
```

This is called a function call.

A function runs only when we call it.


# 2. Calling a Function Multiple Times

A function can be called many times.


````python
def greet():
    print("Hello Student")

greet()
greet()
greet()
````

# 3. Function With Parameter

A parameter is a value that we pass to a function.

It makes the function more useful and flexible.


````python
def greet_student(name):
    print("Hello", name)

greet_student("Ali")
greet_student("Sara")
greet_student("Ahmed")
````

# 4. Function With Multiple Parameters

A function can take more than one parameter.


````python
def student_info(name, age):
    print("Name:", name)
    print("Age:", age)

student_info("Ali", 20)
student_info("Sara", 18)
````

# 5. Function for Addition

Functions are very useful for calculations.


````python
def add_numbers(num1, num2):
    total = num1 + num2
    print("Total:", total)

add_numbers(10, 20)
add_numbers(50, 30)
````

# 6. Function With User Input

We can also take input from the user and pass it to a function.


````python
def check_age(age):
    if age >= 18:
        print("You are eligible")
    else:
        print("You are not eligible")

user_age = int(input("Enter your age: "))
check_age(user_age)
````

# 7. Function With if/else

This example connects functions with control structures.


````python
def check_result(marks):
    if marks >= 50:
        print("Pass")
    else:
        print("Fail")

check_result(80)
check_result(35)
````

# 8. Function That Returns a Value

Sometimes we do not want to print inside a function.

Instead, we want the function to give a value back.

For this, we use `return`.


````python
def add_numbers(num1, num2):
    total = num1 + num2
    return total

result = add_numbers(10, 20)
print("Result:", result)
````

# Difference Between print and return

## print
`print()` only displays the result on screen.

## return
`return` sends the result back so we can store it and use it later.


````python
def square(number):
    return number * number

answer = square(5)
print("Square is:", answer)
````

# 9. Function With List

A function can also work with a list.


````python
def show_marks(marks):
    print("Student Marks:", marks)

student_marks = [80, 75, 90]
show_marks(student_marks)
````

# 10. Function to Calculate Average

This is a practical example using list, variables, and arithmetic operators.


````python
def calculate_average(marks):
    total = sum(marks)
    average = total / len(marks)
    return average

student_marks = [80, 70, 90]
avg = calculate_average(student_marks)

print("Average Marks:", avg)
````

# 11. Function With Dictionary

A dictionary stores data in key-value pairs.

Functions can receive and use dictionaries.


````python
def show_student(student):
    print("Name:", student["name"])
    print("Age:", student["age"])
    print("Course:", student["course"])

student_data = {
    "name": "Ali",
    "age": 20,
    "course": "Python"
}

show_student(student_data)
````

# 12. Function With Default Parameter

A default parameter has a value already set.

If the user does not provide a value, Python uses the default value.


````python
def greet(name="Student"):
    print("Hello", name)

greet("Ali")
greet()
````

# 13. Function With Keyword Arguments

Keyword arguments make the code easier to read.


````python
def student_info(name, age, course):
    print("Name:", name)
    print("Age:", age)
    print("Course:", course)

student_info(name="Ali", age=20, course="Python")
````

# 14. Mini Project: Student Result System Using Function

This example combines:

- input
- variables
- data types
- arithmetic operators
- control structures
- functions


````python
def calculate_result(name, python_marks, english_marks, computer_marks):
    total = python_marks + english_marks + computer_marks
    average = total / 3

    if average >= 80:
        result = "Excellent"
    elif average >= 60:
        result = "Good"
    elif average >= 50:
        result = "Pass"
    else:
        result = "Fail"

    print("\n--- Student Result Card ---")
    print("Name:", name)
    print("Total Marks:", total)
    print("Average Marks:", average)
    print("Result:", result)


name = input("Enter student name: ")
python_marks = int(input("Enter Python marks: "))
english_marks = int(input("Enter English marks: "))
computer_marks = int(input("Enter Computer marks: "))

calculate_result(name, python_marks, english_marks, computer_marks)
````

# 15. Slightly Advanced Example: Banking Function

This example uses a function to perform a banking operation.


````python
def withdraw(balance, amount):
    if amount <= balance:
        balance = balance - amount
        print("Withdrawal Successful")
    else:
        print("Insufficient Balance")

    return balance


current_balance = 5000
withdraw_amount = float(input("Enter withdraw amount: "))

current_balance = withdraw(current_balance, withdraw_amount)

print("Final Balance:", current_balance)
````

# Common Mistakes in Functions

## Mistake 1: Forgetting to call the function

Wrong:
```python
def greet():
    print("Hello")
```

Correct:
```python
def greet():
    print("Hello")

greet()
```

## Mistake 2: Forgetting colon

Wrong:
```python
def greet()
```

Correct:
```python
def greet():
```

## Mistake 3: Wrong indentation

Wrong:
```python
def greet():
print("Hello")
```

Correct:
```python
def greet():
    print("Hello")
```

## Mistake 4: Passing missing arguments

Wrong:
```python
def add(a, b):
    print(a + b)

add(5)
```

Correct:
```python
add(5, 10)
```
