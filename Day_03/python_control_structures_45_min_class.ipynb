# Part 2: What is a Control Structure?

A control structure helps a program make decisions.

Real-life examples:

- If it is raining, take an umbrella.
- If password is correct, login.
- If marks are 50 or above, pass.
- If age is 18 or above, allow entry.

Python can also make decisions using `if`, `else`, and `elif`.


# Visual Understanding

```text
            Condition
           /         \
        True         False
        /              \
   Do this work     Do other work
```

Python checks a condition.

If the condition is true, it runs one block.

If the condition is false, it runs another block.


# Part 3: Guard example

## Real-world idea
A person can enter only if age is 18 or above.

```text
Age >= 18 ?
Yes → Entry Allowed
No  → Not Allowed
```


````python
age = int(input("Enter your age: "))

if age >= 18:
    print("Entry Allowed")
else:
    print("Not Allowed")
````

# Explanation of Syntax

```python
if condition:
    code runs if condition is true
else:
    code runs if condition is false
```

## Important points

- `if` means agar
- `else` means warna
- `:` means block starts here
- Spaces before print are called indentation
- Indentation is very important in Python


# Part 4: Comparison Operators

| Operator | Meaning | Example |
|---|---|---|
| `>` | greater than | `age > 18` |
| `<` | less than | `age < 18` |
| `>=` | greater than or equal | `age >= 18` |
| `<=` | less than or equal | `age <= 18` |
| `==` | equal to | `password == "1234"` |
| `!=` | not equal to | `name != "Ali"` |

## Important
`=` is used to store value.

`==` is used to compare value.


````python
# Difference between = and ==

password = "1234"

print(password >="333")
print(password == "0000")
````

# Part 7: What is elif?

Sometimes we have more than two conditions.

Example:

- Marks 80 or above → Excellent
- Marks 50 or above → Pass
- Otherwise → Fail

For multiple conditions, we use `elif`.


````python
marks = int(input("Enter your marks: "))

if marks >= 80:
    print("Excellent")
elif marks >= 50:
    print("Pass")
else:
    print("Fail")
````

# Visual Understanding of elif


Python checks conditions from top to bottom.


# Part 8: Small Real-World Project — ATM Balance Checker

Take balance from user.

If balance is 1000 or more:

`Withdrawal Allowed`

Otherwise:

`Insufficient Balance`


````python
balance = int(input("Enter your balance: "))

# 2000 is greater then or equal to 1000

if balance >= 1000:
    print("Withdrawal Allowed")
else:
    print("Insufficient Balance")
````

# Part 9: Better ATM Program

Now take withdrawal amount also.

If balance is greater than or equal to withdrawal amount, allow withdrawal.

Otherwise, show insufficient balance.


````python
balance = int(input("Enter your balance: "))
withdraw_amount = int(input("Enter withdrawal amount: "))




if balance >= withdraw_amount:
    print("Withdrawal Successful")
else:
    print("Insufficient Balance")
````

# Common Mistakes

## Mistake 1: Using = instead of ==

Wrong:
```python
if password = "1234":
```

Correct:
```python
if password == "1234":
```

## Mistake 2: Forgetting colon

Wrong:
```python
if age >= 18
```

Correct:
```python
if age >= 18:
```

## Mistake 3: Wrong indentation

Wrong:
```python
if age >= 18:
print("Allowed")
```

Correct:
```python
if age >= 18:
    print("Allowed")
```
