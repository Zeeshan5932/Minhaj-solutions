# Arithmetic Operators

## What are Arithmetic Operators?

Arithmetic operators are used to perform **basic mathematical operations** in programming.

Simple words:
They help us **calculate values**, just like we do in daily life.


## Why do we use Arithmetic Operators?

We use them to:

* Calculate total marks
* Find average
* Do billing calculations
* Perform daily math operations inside programs

👉 Without arithmetic operators, programs cannot **calculate anything**.



## Real-Life Story (Simple Understanding)

Imagine a student **Ali**:

* He got marks in 3 subjects
* Now he wants to find his **total marks** and **average**

👉 He will:

* Add marks → Total
* Divide total → Average

This is exactly what arithmetic operators do in programming.

---

## Types of Arithmetic Operators

| Operator | Name           | What it does (Simple)                   | Real-Life Example               |
| -------- | -------------- | --------------------------------------- | ------------------------------- |
| +        | Addition       | Do values ko add karta hai              | 70 + 80 = 150 (Total marks)     |
| -        | Subtraction    | Difference nikalta hai                  | 100 - 20 = 80 (Remaining money) |
| *        | Multiplication | Values ko multiply karta hai            | 5 × 200 = 1000 (Total price)    |
| /        | Division       | Value ko divide karta hai               | 150 ÷ 3 = 50 (Average marks)    |
| %        | Modulus        | Remainder deta hai                      | 7 ÷ 2 = remainder 1             |
| **       | Power          | Number ko power deta hai                | 2² = 4                          |
| //       | Floor Division | Sirf poora number deta hai (no decimal) | 7 ÷ 2 = 3                       |

---


## Important Point

Arithmetic operators are the **base of programming logic**.

👉 Every real system uses them:

* Banking systems
* Shopping websites
* Result systems
* Salary calculations

## 2. Lists

Store multiple values in one variable.(mutable)

Example: marks list, shopping list



## Why we use List?

* To store **multiple values** in one variable
* To avoid creating many variables
* To manage data easily (like marks, names, items)

---



## List Properties

| Feature         | Description       |
| --------------- | ----------------- |
| Ordered         | Yes (index based) |
| Changeable      | Yes (mutable)     |
| Allow Duplicate | Yes               |
| Data Types      | Multiple allowed  |



## Key Point

List = **Multiple values in one container**


**Integer List (Marks System)**

````python
marks = [80, 75, 90, 60]

print("All Marks:", marks)
print("First Student Marks:", marks[3])
print("lowest Marks:", min(marks))
````

**Float List (Temperature Tracker)**

````python
temperatures = ([36.5, 37.2, 38.0, 23.4])

print("Temperatures:", temperatures)
print("Highest Temperature:", max(temperatures))
print("Lowest Temperature:", min(temperatures))
print("sum Temperature:", sum(temperatures))
print("Average Temperature:", sum(temperatures) / len(temperatures))
````

**String List (Friends List)**

````python
game_results = [True, False, True, True]

print("Game Results:", game_results)
print("First Game Win?", game_results[2])
print("Total Wins:", game_results.count(True))
````

## 3. Tuples

Same as list but cannot be changed.




### What is Tuple?

A **tuple** is a collection of multiple values stored in a single variable.
It is **ordered** and **immutable (cannot be changed after creation)**.


## Why We Use Tuple?

* To store **fixed data** (that should not change)
* Faster than lists in some cases
* Used when data needs to be **safe (read-only)**

---

## Tuple vs List (Quick Table)

| Feature     | Tuple `()`       | List `[]`       |
| ----------- | ---------------- | --------------- |
| Brackets    | `()`             | `[]`            |
| Changeable  | ❌ No (Immutable) | ✅ Yes (Mutable) |
| Order       | ✅ Yes            | ✅ Yes           |
| Performance | ⚡ Faster         | ⏳ Slower        |

---

# 🎯 ✅ 1. LIST Example (marks + result)

```python
# List + input + control structure

marks = [
    int(input("Enter Python marks: ")),
    int(input("Enter English marks: ")),
    int(input("Enter Computer marks: "))
]

total = sum(marks)
average = total / 3

if average >= 50:
    print("Pass")
else:
    print("Fail")

print("Marks List:", marks)
print("Average:", average)
```

---

# 🎯 ✅ 2. TUPLE Example (fixed subjects)

```python
# Tuple + control structure

subjects = ("Python", "English", "Computer")

marks = (
    int(input("Enter Python marks: ")),
    int(input("Enter English marks: ")),
    int(input("Enter Computer marks: "))
)

if marks[0] >= 50 and marks[1] >= 50 and marks[2] >= 50:
    print("All subjects passed")
else:
    print("Some subjects failed")

print("Subjects:", subjects)
print("Marks:", marks)
```

---

# 🎯 ✅ 3. DICTIONARY Example (student profile)

```python
# Dictionary + input + control structure

student = {
    "name": input("Enter name: "),
    "age": int(input("Enter age: ")),
    "marks": int(input("Enter marks: "))
}

if student["marks"] >= 50:
    result = "Pass"
else:
    result = "Fail"

print("\nStudent Info")
print("Name:", student["name"])
print("Age:", student["age"])
print("Marks:", student["marks"])
print("Result:", result)
```

# 🎯 ✅ 4. SET Example (unique values + condition)

```python
# Set + input + control structure

nums = {
    int(input("Enter number 1: ")),
    int(input("Enter number 2: ")),
    int(input("Enter number 3: "))
}

if len(nums) < 3:
    print("Duplicate values entered")
else:
    print("All values are unique")
```
