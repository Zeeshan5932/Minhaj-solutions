# Python Loops — Beginner to Advanced

## What are Loops?
Loops repeat a block of code.

## Why use loops?
- Avoid repetition
- Handle multiple values
- Automate tasks

## Types:
- for loop
- while loop


# FOR LOOP EXAMPLES

## Example 1: Print Hello

```python
for i in range(5):
    print('Hello')
```

## Example 2: Numbers 1 to 5

```python
for i in range(1,6):
    print(i)
```

## Example 3: Even numbers

```python
for i in range(1,11):
    if i%2==0:
        print(i)
```

## Example 4: Reverse

```python
for i in range(5,0,-1):
    print(i)
```

## Example 5: List loop

```python
marks=[80,70,90]
for m in marks:
    print(m)
```

## Example 6: Sum

```python
total=0
for i in range(1,6):
    total+=i
print(total)
```

## Example 7: Nested loop

```python
for i in range(3):
    for j in range(3):
        print('*',end=' ')
    print()
```

## Example 8: String loop

```python
for ch in 'Python':
    print(ch)
```

## Example 9: Range step

```python
for i in range(0,10,2):
    print(i)
```

# WHILE LOOP EXAMPLES

## Example 1

```python
i=1
while i<=5:
    print(i)
    i+=1
```

## Example 2 Reverse

```python
i=5
while i>=1:
    print(i)
    i-=1
```

## Example 3 Sum

```python
i=1
total=0
while i<=5:
    total+=i
    i+=1
print(total)
```

## Example 4 Password

```python
password=''
while password!='1234':
    password=input('Enter password: ')
print('Access Granted')
```

## Example 5 Break

```python
i=1
while i<=10:
    if i==5:
        break
    print(i)
    i+=1
```

# Mini Project

```python
marks=[]
for i in range(5):
    m=int(input('Enter marks: '))
    marks.append(m)

total=0
for m in marks:
    total+=m

print('Total:',total)
print('Average:',total/len(marks))
```
