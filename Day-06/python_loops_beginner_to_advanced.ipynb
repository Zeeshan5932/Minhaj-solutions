# Python Loops — Beginner to Advanced

## What is a Loop?
A loop is used to repeat a block of code multiple times.

## Why do we use loops?
- To avoid writing repeated code
- To automate tasks
- To work with multiple values

---

## FOR LOOP
Used when we know how many times to repeat.


```python
# Basic for loop
for i in range(5):
    print("Hello")
```

### Explanation
range(5) means loop runs 5 times.


```python
# Print numbers 1 to 5
for i in range(1,6):
    print(i)
```

```python
# Print even numbers
for i in range(2,11,2):
    print(i)
```

```python
# Loop with list
marks = [80,70,90]
for m in marks:
    print(m)
```

```python
# Pass/Fail using loop
marks = [80,40,60]
for m in marks:
    if m >= 50:
        print("Pass")
    else:
        print("Fail")
```

---
## WHILE LOOP
Used when condition is given.


```python
# Basic while loop
i = 1
while i <= 5:
    print(i)
    i = i + 1
```

```python
# Sum of numbers
i = 1
total = 0

while i <= 5:
    total += i
    i += 1

print("Total:", total)
```

```python
# Reverse counting
i = 5
while i >= 1:
    print(i)
    i -= 1
```

```python
# Infinite loop example (avoid)
# while True:
#     print("Hello")
```

---
## Difference Between for and while

FOR LOOP:
- Fixed repetition
- Easy to use

WHILE LOOP:
- Condition based
- Flexible but risky

## Advanced Example

Find even numbers from list


```python
numbers = [1,2,3,4,5,6,7,8]

for n in numbers:
    if n % 2 == 0:
        print(n)
```

**Even & Odd Counter**

```python
numbers = [10, 15, 22, 33, 40]

even = 0
odd = 0

for num in numbers:
    if num % 2 == 0:
        even += 1
    else:
        odd += 1

print("Even:", even)
print("Odd:", odd)
```

## Password Retry System

```python
password = "1234"

while True:
    user = input("Enter password: ")

    if user == password:
        print("Access Granted")
        break
    else:
        print("Wrong password, try again")
```

## Multiplication Table Generator

```python
num = int(input("Enter number: "))

for i in range(1, 11):
    print(num, "x", i, "=", num * i)
    
```

```python
balance = 1000

while True:
    print("\n1. Check Balance")
    print("2. Deposit")
    print("3. Exit")

    choice = input("Enter choice: ")

    if choice == "1":
        print("Balance:", balance)

    elif choice == "2":
        amount = int(input("Enter amount: "))
        balance += amount
        print("Deposited!")

    elif choice == "3":
        print("Goodbye!")
        break

    else:
        print("Invalid choice")
```
