## What are Data Types?

In Python, every value has a type. A data type tells Python what kind of value is stored in a variable.

For example:

- Text data is stored as `str`
- Whole numbers are stored as `int`
- Decimal numbers are stored as `float`
- True/False values are stored as `bool`

Data types help Python understand what operations can be performed on a value.

```python
name = (20)
age = (20)
height = 5.8
is_student = True # boolen

print(name)
print(age)
print(height)
print(is_student)
```

```python
name = "Ali"
second_name = "Ahmed"
print("Full name is :", name + " \n" +second_name)
```

## Why Do We Use Data Types?

We use data types because different kinds of data behave differently.

For example:

- Numbers can be added, subtracted, multiplied, and divided
- Text can be joined with other text
- Boolean values are used for decisions
- Lists can store multiple values in one variable

Data types make programs more organized, clear, and logical.

## Checking Data Type with `type()`

Python provides the `type()` function to check the data type of any value or variable.

```python
name = "Ali"
age = 20.0
height = 5.8
is_student = True

print(type(name))
print(type(age))
print(type(height))
print(type(is_student))
```

# 1. String Data Type `str`

A string is used to store text data.

Strings are written inside single quotes `' '` or double quotes `" "`.

Examples of strings:

- Name
- City
- Email
- Address
- Message

```python
student_name = "Ahmed"
city = "Lahore"
message = "Welcome to Python class"

print(student_name)
print(city)
print(message)

print(type(student_name))
```

## String Operations

Strings can be joined together using `+`. This is called string concatenation.

```python
first_name = "Ali"
last_name = "Khan"

full_name = first_name + " " + last_name

print(full_name)
```

## String Methods

Python has built-in functions for strings. These are called string methods.

```python
text = "Python is easy to undertsand language"

print(text.upper())
print(text.lower())
print(text.title())
print(len(text))
```

# 2. Integer Data Type `int`

An integer is used to store whole numbers without decimal points.

Examples:

- Age
- Marks
- Number of students
- Quantity of products

```python
age = 18
marks = 85
students = 30

print(age)
print(marks)
print(students)

print(type(age))
```

## Integer Operations

Integers can be used in mathematical operations.

```python
a = 10
b = 3

print("Addition:", a + b)
print("Subtraction:", a - b)
print("Multiplication:", a * b)
print("Division:", a / b)
print("Floor Division:", a // b)
print("Modulus:", a % b)
print("Power:", a ** b)
```

# 3. Float Data Type `float`

A float is used to store decimal numbers.

Examples:

- Height
- Weight
- Temperature
- Price
- Percentage

```python
height = 5.8
weight = 62.5
temperature = 37.2
price = 499.99

print(height)
print(weight)
print(temperature)
print(price)

print(type(price))
```

## Float Calculations

Floats are commonly used when exact decimal values are needed.

```python
price = 1200.50
discount = 200.25

final_price = price - discount

print("Final Price:", final_price)
```

# 4. Boolean Data Type `bool`

A boolean stores only two possible values:

- `True`
- `False`

Boolean values are mostly used in decision making.

Examples:

- Is user logged in?
- Is student passed?
- Is payment completed?
- Is product available?

```python
is_logged_in = True
is_passed = False
is_available = True

print(is_logged_in)
print(is_passed)
print(is_available)

print(type(is_logged_in))
```

# 5. List Data Type `list`

A list is used to store multiple values in one variable.

Lists are written inside square brackets `[ ]`.

A list can store:

- Multiple names
- Multiple marks
- Multiple products
- Mixed data types

```python
students = ["Ali", "Ahmed", "Sara", "Zain"]
marks = [80, 75, 90, 88]
mixed_data = ["Ali", 20, 5.8, True]

print(students)
print(marks)
print(mixed_data)

print(type(students))
```

## Accessing List Items

List indexing starts from `0`.

```python
fruits = ["Apple", "Banana", "Mango", "Orange"]

print(fruits[0])
print(fruits[1])
print(fruits[2])
print(fruits[-1])
```

# 6. Tuple Data Type `tuple`

A tuple is also used to store multiple values.

Tuples are written inside round brackets `( )`.

The main difference is that tuples cannot be changed after creation.

This means tuples are immutable.

```python
colors = ("Red", "Green", "Blue")

print(colors)
print(colors[0])
print(type(colors))
```

# 7. Dictionary Data Type `dict`

A dictionary stores data in key-value pairs.

Dictionaries are written inside curly brackets `{ }`.

Each item has:

- Key
- Value

Example:

`"name" : "Ali"`

```python
student = {
    "name": "Ali",
    "age": 20,
    "course": "Python",
    "is_enrolled": True
}

print(student)
print(type(student))
```

## Accessing Dictionary Values

Values can be accessed using their keys.

```python
student = {
    "name": "Ali",
    "age": 20,
    "course": "Python"
}

print(student["name"])
print(student["age"])
print(student["course"])
```

# 8. Set Data Type `set`

A set is used to store unique values.

Sets are written inside curly brackets `{ }`.

Sets automatically remove duplicate values.

```python
numbers = {1, 2, 3, 3, 4, 4, 5}

print(numbers)
print(type(numbers))
```
